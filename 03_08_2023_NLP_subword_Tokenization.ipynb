```python
train_data = imdb_plaintext['train']

train_sentences = []

for s, l in train_data:
  train_sentences.append(s.numpy().decode('utf-8'))
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
vocab_size = 10000
oov_tok = '<oov>'

tokenizer_plaintext = Tokenizer(num_words=vocab_size,
                                oov_token=oov_tok)

tokenizer_plaintext.fit_on_texts(train_sentences)
```

```python
sequences_train = tokenizer_plaintext.texts_to_sequences(train_sentences)
```

```python
print(sequences_train[:2])
```

```python
tokenizer_plaintext.sequences_to_texts(sequences_train[:2])
```
